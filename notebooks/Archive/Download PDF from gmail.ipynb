```python
import email
import os
import imaplib
import base64
import tabula
import pandas as pd
```

```python
email_user = 'ajitisack@gmail.com'
email_pass = 'rigid.03'
host = 'imap.gmail.com'
port = 993
mail = imaplib.IMAP4_SSL(host, port)
mail.login(email_user, email_pass)
mail.select('Trading')
```

```python
dfs = []
result, data = mail.search(None, '(FROM "NSE-Direct")' )
email_ids = data[0].split() # data is a list.
for email_id in email_ids:
    result, data = mail.fetch(email_id, "(RFC822)")
    email_string = data[0][1].decode('utf-8')
    email_message = email.message_from_string(email_string)
    for part in email_message.walk():
        fileName = part.get_filename()
        if fileName: 
#             print(fileName)
            filePath = f'/Users/ajit/Downloads/{fileName}'
            fp = open(filePath, 'wb')
            fp.write(part.get_payload(decode=True))
            fp.close()
            tables = tabula.read_pdf(filePath, pages = "all", multiple_tables = True, password='ABEPI5291E')
            dfs.append(pd.concat(tables))
```

```python
df = pd.concat(dfs, ignore_index=True)
df = df[df.Client != 'Code'].reset_index(drop=True)
```
