```python
import stockdata as sd
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
sd.getdata("select * from symbols where name like 'Jubilan%'")
```

```python
def nsechart(symbol, column='close'):
    df = sd.gethistprice('NSE', symbol)
    df.set_index('date', inplace=True)
    df[column].plot(figsize=(20, 8))
    plt.title(f"{symbol} - Close Price", fontsize=20)
    plt.ylabel('Price', fontsize=14)
    plt.xlabel('Date', fontsize=14)
    plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
    plt.show()
```

```python
nsechart('JUBILANT')
```

```python
def nsechart_indexsymbols(index_name):
    symbols = sd.getdata(f"select exchange, symbol from nseindices where indexname = '{index_name}'").symbol.to_list()
    [nsechart(i) for i in symbols]
```

```python
nsechart_indexsymbols('Nifty Bank')
```
